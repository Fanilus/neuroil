```
import pandas as pd
from datetime import datetime
from datetime import timedelta
```

```
producted = pd.read_csv('producted.csv')
sold = pd.read_csv('sold.csv')
```

```
print(producted.keys())
```

```
str_to_print = ''
for j in producted.keys():
  if j == 'date': continue
  str_to_print += "'["
  for i in range(len(producted)):
    str_to_print += f"[{producted['date'][i]}, {str(int(producted[j][i])-int(sold[j][i]))}], "
  str_to_print = str_to_print[:-2] + "]'\n"
print(str_to_print)
```
